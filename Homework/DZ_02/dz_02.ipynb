#### Задание 3

```python
# Создание массива x
x = np.linspace(-3, 3, 51)

# Создание массивов y1, y2, y3, y4
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
# Создаем объект Figure и массив Axes
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))

# Устанавливаем зазоры между графиками
plt.subplots_adjust(wspace=0.3, hspace=0.3)

# Построение графиков
ax[0, 0].plot(x, y1)
ax[0, 0].set_title("График y1")
ax[0, 0].set_xlim(-5, 5)

ax[0, 1].plot(x, y2)
ax[0, 1].set_title("График y2")

ax[1, 0].plot(x, y3)
ax[1, 0].set_title("График y3")

ax[1, 1].plot(x, y4)
ax[1, 1].set_title("График y4")

# Отображаем графики
plt.show()
```

#### Задание 4

```python
# Загрузка датасета
df = pd.read_csv("creditcard.csv")
```

```python
# Подсчет количества наблюдений для каждого значения целевой переменной Class
class_counts = df["Class"].value_counts()
class_counts
```

```python
# Установка стиля для графиков
plt.style.use("fivethirtyeight")
```

```python
# Построение столбчатой диаграммы
class_counts.plot(kind="bar")
plt.title("Количество наблюдений для каждого класса")
plt.xlabel("Класс")
plt.ylabel("Количество")
plt.show()
```

```python
# Построение столбчатой диаграммы с логарифмическим масштабом
class_counts.plot(kind="bar", logy=True)
plt.title("Количество наблюдений для каждого класса (логарифмический масштаб)")
plt.xlabel("Класс")
plt.ylabel("Логарифм количества")
plt.show()
```

```python
# Построение гистограмм для V1 для каждого класса
plt.hist(df[df["Class"] == 0]["V1"], bins=20, alpha=0.5, density=True, label="Class 0", color="gray")
plt.hist(df[df["Class"] == 1]["V1"], bins=20, alpha=0.5, density=True, label="Class 1", color="red")
plt.title("Распределение V1 для каждого класса")
plt.xlabel("V1")
plt.ylabel("Плотность")
plt.legend()
plt.show()
```

```python
a = np.arange(12)
A = a.reshape(4, 3)
At = A.T
B = np.dot(A, At)

print("Размер матрицы B:", B.shape)

# Пытаемся вычислить обратную матрицу B
try:
    B_inv = np.linalg.inv(B)
    print("Обратная матрица B:\n", B_inv)
except np.linalg.LinAlgError:
    print("Невозможно вычислить обратную матрицу для B, так как она не является квадратной (количество строк не равно количеству столбцов).")
```

```python
np.random.seed(42)
```

```python
c = np.random.randint(0, 16, 16)
c
```

```python
C = c.reshape(4, 4)
D = C * 10 + B
det_D = np.linalg.det(D)
print("Определитель матрицы D:", det_D)

rank_D = np.linalg.matrix_rank(D)
print("Ранг матрицы D:", rank_D)

try:
    D_inv = np.linalg.inv(D)
    print("Обратная матрица D:\n", D_inv)
except np.linalg.LinAlgError:
    print("Невозможно вычислить обратную матрицу для D.")
```

```python
D_inv_binary = np.where(D_inv > 0, 1, 0)
is_binary = np.array_equal(D_inv_binary, D_inv_binary.astype(bool))
print(is_binary)

E = np.where(D_inv_binary == 1, B, C)
print("Матрица E:\n", E)
```
